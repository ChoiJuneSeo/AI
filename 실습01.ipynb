```python
import pandas as pd
import numpy as np
```

```python
nbastat = pd.read_csv('nbastat2022.csv')
nbastat
```

```python
nbastat2 = nbastat[['FGA','FGM']]
m = len(nbastat)
m
```

```python
X = nbastat2[['FGA']]
Y = nbastat2[['FGM']]
```

```python
X = X.fillna(0)
Y = Y.fillna(0)
```

```python
X = (np.array(X)).reshape(m,1)
Y = np.array(Y).reshape(m,1)
```

```python
import matplotlib.pyplot as plt
plt.plot(X,Y,'.b')
```

```python
x0 = np.ones((m,1))
Xb = np.c_[x0,X]
n_iter = 20000
theta = np.zeros((2,1))
learning_rate = 0.001
gradients = np.zeros((2,1))
```

```python
for i in range(n_iter):
  gradients = (1./m)*(Xb.T.dot(Xb.dot(theta)-Y))
  theta = theta - learning_rate * gradients
theta
```

```python
Y_pred = Xb.dot(theta)
plt.plot(X, Y_pred,color='Red')
plt.plot(X,Y,'.b')
```

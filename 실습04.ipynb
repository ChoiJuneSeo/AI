```python
import numpy as np

def euclidean_distance(x1, x2):
  return np.sqrt(np.sum((x1-x2)**2))

def kmeans(X,k,max_iters):
  m, n = X.shape

  centroid_id = np.random.choice(m,k,replace=False)
  centroids = X[centroid_id];

  for i in range(max_iters):
    # 1. 모든 포인트에서 모든 centroids와의 거리를 계산해서 가장 가까운 centroids로 cluster
    distances = []
    for s in range(m):
      distance = []
      for j in range(k):
        distance.append(euclidean_distance(X[s],centroids[j]))
      distances.append(distance)
    label = np.argmin(distances,axis=1)
    # 2. 동일한 cluster에 속한 모든 point에 대한 평균을 계산해서 새로운 centroid로 결정
    new_centroids = np.array([X[label == s].mean(axis=0) for s in range(k)])

    if(np.array_equal(centroids,new_centroids)):
      print(f"{i}번")
      break;

    centroids = new_centroids

  return label




```
